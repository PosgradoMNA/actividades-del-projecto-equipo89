**Información del Conjunto de Datos**

Airbnb Open Data (New York)

Publicado por **Airbnb Inside**, organización sin fines de lucro que analiza el impacto socioeconómico que AirbnB ejerce en las comunidades. Dataset modificado por Arian Azmoudeh con el motivo de praticar limpieza de los datos.

[Enlace al dataset original.](http://insideairbnb.com/explore/)

[Enlace al dataset a utilizar.](https://www.kaggle.com/datasets/arianazmoudeh/airbnbopendata)

---

El dataset seleccionado consta de una lista de alojamientos disponibles para renta en Airbnb. Lista creada con el motivo de comprobar que una gran mayoría dichas rentas se tratan de propiedades que solo se pueden rentar, uno, por completo y, dos, mensualmente. Es decir, que la mayoría de los alojamientos se tratan de una renta inmboliria típica, pero con una sobre tarifa de renta. Lo que, inevitablemente, conlleva a una injustificada alza de precio sobre la vivienda, entre otras circunstancias desfavorables para la comunidad.


---
**Preguntas de Interés**
1. ¿Qué porcentaje de los alojamientos se tratan de propiedades completas?
2. ¿Qué porcentaje de los alojamientos se tratan de propiedades completas, que solo se pueden rentar mensualmente?
3. ¿Qué vecindarios de Nueva York tienen una mayor tasa del segundo tipo de alojamientos?
4. ¿Qué vecindarios de Nueva York tienen una menor tasa del segundo tipo de alojamientos?
5. ¿Cuál es la tasa de disponibilidad de este tipo de alojamientos?
6. Considerando el tiempo de disponibilidad de los alojamientos, ¿qué tipo de alojamiento genera mayores ingresos para Airbnb según su tarifa de servicio?
7. ¿Cuáles son las características de los alojamientos que tienen una puntuación de más de 4 estrellas?
8. ¿Afecta el tipo de reseñas (positivas y negativas) y la cantidad de reseñas en el precio del Airbnb o de la tasa de servicio que se ofrece?



---
Como lo indica Airbnb Inside, los interesados en esta información son los propios habitantes de la comunidad en cuestión. Es decir, este dataset en concreto sería de utilidad para Sociedades Civiles del estado de Nueva York, así como para interesados en rentar algún inmueble o alojamiento. De igual manera, podría ser de utilidad para cualquier persona que quiera incursionar en los servicios de AirBnB para poder ofrecer un servicio de buena calidad
