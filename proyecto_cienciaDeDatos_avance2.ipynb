# Avance 2 del proyecto

Para nuestro proyecto, se ha seleccionado la metodología Cross-Industry Standard Process for Data Mining (CRISP-DM). La cual consta de cinco fases: 

1. Comprensión del Negocio
1. Comprensión de los Datos
1. Preparación de los Datos
1. Modelado
1. Evaluación
1. y Despliegue

A continuación, se detalla cada fase en cuestión.


**1. Comprensión del Negocio**

Se comienza por definir los objetivos y el alcance del proyecto en cuestiones de negocio. Y a partir de ello, se define un problema a resolverse con mineria de datos.

---
**2. Comprensión de los Datos**

Se reciben los datos para familiarizarse con ellos. Se identifican problemas de calidad. Asimismo que se infiere conocimiento preeliminar.

---
**3. Preparación de los Datos**

Consiste en la transformación de los datos en bruto. *'Las tareas incluyen la selección de tablas, registros y atributos, así como la transformación y la limpieza de datos para las herramientas que modelan'.* (Sngular, 2022)

---
**4. Modelado**

Para un mismo problema de mineria de datos, se pueden aplicar más de una técnica de modelado. Por lo que se aplican todas las técnicas posibles en el conjunto. Siempre con el motivo de obtener modelos que ayuden al análisis de datos. 

---
**5. Evaluación**

Una vez se tienen los modelos de datos construidos, se debe evaluar su valor para el negocio. Es decir, se debe aprobar su pertinencia a los objetivos y el alcance definido.

---
**6. Despliegue**

La puesta en producción varia de organización a organización. Sin embargo, siempre en consiste en colocar al alcance la información para las personas de interés. En algunas situaciones, es importante considerar posibles automatizaciones para reiterar la creación de este conocimiento.
